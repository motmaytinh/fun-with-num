```python
import pandas as pd
import numpy as np

from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("data.csv", delim_whitespace=True, header=None, names=['roman_num'] + ['num']+['digit' + str(i) for i in range(17)])
df.head(9)
```

```python
train, test = train_test_split(df, test_size=0.2, random_state=42)

Y_train = train.iloc[:,1]
X_train = train.drop(columns=['roman_num','num'])/1000

Y_test = test.iloc[:,0:2]
X_test = test.drop(columns=['roman_num','num'])/1000
```

```python
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
Y_pred = svr_rbf.fit(X_train, Y_train).predict(X_test)
```

```python
df2 = pd.DataFrame(Y_pred)
df3 = pd.DataFrame(Y_test)
df3.assign(predict = pd.Series(Y_pred, index=df3.index))
```
